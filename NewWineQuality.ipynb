```python
X = df[['type','fixed acidity','volatile acidity','citric acid',
             'residual sugar','chlorides','free sulfur dioxide'	,
             'total sulfur dioxide','density','pH','sulphates','alcohol',
             'quality']].values
X[0:12]
```

```python
#Mengisi variabel target
y = df["quality"]

y[0:12]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#4 parameter data yang mewakili rasio dari data set pengujian, dan random state memastikan mendapatkan pemisahan yang sama
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
WineTree = DecisionTreeClassifier(criterion="gini", max_depth = 10)
WineTree
```

# MODELLING

```python
WineTree.fit(X_trainset,y_trainset)
```

```python
predTree = WineTree.predict(X_testset)
```

# PREDICTION

```python
print (predTree [0:12])
print (y_testset [0:12])
```

```python
#Persentase keberhasilan
WineTree.score(X_testset, y_testset)
```

# EVALUATION

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
